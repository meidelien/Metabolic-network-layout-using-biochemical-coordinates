```python

import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

net2 = nx.DiGraph()

# set the physics layout of the network
# data = pd.read_csv("TCA_target_standardized.csv")
data = pd.read_csv("Chemical_descriptors_e_coli_core_ranked.csv")
target = pd.read_csv("Chemical_descriptors_e_coli_core_ranked_targets.csv")

sources = data["BiGG"]
targets = target['target']
coordinates = data[["massRank", "MolLogPRank"]].values

file_name = "massRank_MolLogPRank"

edge_data = zip(sources,targets, coordinates) # targets

for e in edge_data:
    src = e[0]
    dst = e[1]
    c = e[2]

    net2.add_node(src, pos = c)
    net2.add_edge(src,dst)
    net2.add_edge(dst, src)
    
    
    #net2.add_node(dst, pos = c)
    #net2.add_edge(src, dst)






pos=nx.get_node_attributes(net2,'pos')
fig, ax = plt.subplots(figsize=(10,10))

net2.remove_edges_from(nx.selfloop_edges(net2))

nx.draw(net2,pos, with_labels = True, arrows = True, ax = ax)

limits=plt.axis('on') # turns on axis
ax.set_xlabel("massRank")
ax.set_ylabel("MolLogPRank")
ax.tick_params(left=True, bottom=True, labelleft=True, labelbottom=True)


```

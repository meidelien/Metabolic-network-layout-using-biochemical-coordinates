## Geometry Optimization for `rdkit`

### Demonstrates the capacity to optimize the geometry of a molecule.
- Remember to have PyMol running in the background

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem.Draw.MolDrawing import MolDrawing, DrawingOptions
```

```python
ibu = Chem.MolFromSmiles('CC(C)Cc1ccc(cc1)C(C)C(=O)O')
AMP = Chem.MolFromSmiles("C1=NC(=C2C(=N1)N(C=N2)[C@H]3[C@@H]([C@@H]([C@H](O3)COP(=O)(O)O)O)O)N")
GMP = Chem.MolFromSmiles("C1=NC2=C(N1[C@H]3[C@@H]([C@@H]([C@H](O3)COP(=O)(O)O)O)O)N=C(NC2=O)N")
CMP = Chem.MolFromSmiles("C1=CN(C(=O)N=C1N)[C@H]2[C@@H]([C@@H]([C@H](O2)COP(=O)(O)O)O)O")
Citrate = Chem.MolFromSmiles("C(C(=O)[O-])C(CC(=O)[O-])(C(=O)[O-])O")
Histidine = Chem.MolFromSmiles("C1=C(NC=N1)C[C@@H](C(=O)O)N")
```

```python
ibuH = Chem.AddHs(ibu)
```

```python
AllChem.EmbedMolecule(ibuH)
```

```python
from rdkit.Chem import PyMol
```

```python
v = PyMol.MolViewer()
```

```python
v.ShowMol(ibuH)
```

```python
v.server.do('ray')
v.GetPNG()
```

```python
AllChem.MMFFOptimizeMolecule(ibuH)
```

```python
v.ShowMol(ibuH,name='optimized',showOnly=True);
v.server.do('ray')
v.GetPNG()
```
